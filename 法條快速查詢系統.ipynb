```python
import csv
import pandas as pd
import os
import chardet
```

```python
# 先以單一檔案為測試，插入一列，該列的內容為"法條名稱"，並將標題行改成"法條名稱"、"法條編號"、"法條內容"

# 獲取.ipynb檔案的目錄路徑
notebook_path = os.path.abspath('')
csv_filename = 'Test.csv'  # 替換為您的CSV檔案名稱

# 組合CSV檔案的完整路徑
file_path = os.path.join(notebook_path, csv_filename)

# 使用 chardet 檢測檔案編碼
with open(file_path, 'rb') as f:
    result = chardet.detect(f.read())

# 獲取檢測結果的編碼
encoding = result['encoding']

# 讀取CSV檔案，指定檔案編碼
df = pd.read_csv(file_path, encoding=encoding)

# 複製標題行文字到第一行
df.loc[-1] = df.columns
df.index = df.index + 1
df = df.sort_index()

# 將標題行設為"法條編號"、"法條文字"
df.columns = ['法條編號', '法條內容']

# 插入一列空白的"法條名稱"
df.insert(0, '法條名稱', '')

# 將"法條名稱"列的內容設為檔案名稱（去除附檔名）
df['法條名稱'] = os.path.splitext(csv_filename)[0]

# 將標題行改為"法條名稱"、"法條編號"、"法條內容"
df.columns = ['法條名稱', '法條編號', '法條內容']

# 顯示處理後的資料框
df.head()

# 如果需要保存更改，可以使用以下代碼
df.to_csv('Test.csv', index=False, encoding='utf-8-sig')  # 替換為您想要保存的檔案名稱
```

```python
# 獲取.ipynb檔案的目錄路徑
notebook_path = os.path.abspath('')

# 取得所有CSV檔案的檔案名稱列表
csv_files = [file for file in os.listdir(notebook_path) if file.endswith('.csv')]

for csv_file in csv_files:
    # 使用 chardet 檢測檔案編碼
    with open(os.path.join(notebook_path, csv_file), 'rb') as f:
        result = chardet.detect(f.read())

    # 獲取檢測結果的編碼
    encoding = result['encoding']

    # 讀取CSV檔案，指定檔案編碼
    df = pd.read_csv(os.path.join(notebook_path, csv_file), encoding=encoding)

    # 複製標題行文字到第一行
    df.loc[-1] = df.columns
    df.index = df.index + 1
    df = df.sort_index()

    # 將標題行設為"法條編號"、"法條文字"
    df.columns = ['法條編號', '法條內容']

    # 插入一列空白的"法條名稱"
    df.insert(0, '法條名稱', '')

    # 將"法條名稱"列的內容設為檔案名稱（去除附檔名）
    df['法條名稱'] = os.path.splitext(os.path.basename(csv_file))[0]

    # 將標題行改為"法條名稱"、"法條編號"、"法條內容"
    df.columns = ['法條名稱', '法條編號', '法條內容']

    # 顯示處理後的資料框
    print(df.head())

    # 如果需要保存更改，可以使用以下代碼
    df.to_csv(os.path.join(notebook_path, csv_file), index=False, encoding='utf-8-sig')
```

```python
#合併所有CSV檔

# 取得目前工作目錄
current_directory = os.getcwd()

# 列出目錄中的所有檔案
files = [file for file in os.listdir(current_directory) if file.endswith('.csv')]

# 預期的標題行
expected_header = ['法條名稱', '法條編號', '法條內容']

# 儲存所有 CSV 檔案的資料框架的列表
dfs = []

# 迴圈處理每個 CSV 檔案
for file_name in files:
    # 開啟檔案，讀取資料框架
    df = pd.read_csv(file_name, delimiter='\t', encoding='utf-8')

    # 檢查標題行是否符合預期
    actual_header = list(df.columns)
    if actual_header == expected_header:
        print(f'檔案 {file_name} 的標題行不符合預期。')
    else:
        # 將資料框架加入列表
        dfs.append(df)

# 合併所有資料框架
result_df = pd.concat(dfs, ignore_index=True)

# 將合併後的結果寫入新的 CSV 檔案
result_df.to_csv('merged_file.csv', index=False, sep='\t', encoding='utf-8-sig')

print('合併完成，結果已寫入 merged_file.csv。')
```

```python
# 讀取CSV檔案
df = pd.read_csv('merged_file.csv')

# 使用正則表達式提取包含一、二、三等數字並以"元"結尾的文字，新增一列"處罰內容"
df['處罰內容'] = df['法條內容'].str.extract(r'([一二三四五六七八九十百千萬]+元)', expand=False)

# 顯示處理後的DataFrame
print(df['處罰內容'])

# 將處理後的DataFrame寫回CSV檔案
df.to_csv('output_file.csv', index=False, encoding='utf-8-sig')
```
